```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('C:\\Users\\admin\\Desktop\\stage\\The sparks foundation\\task2\\color-identification-in-image-main\\color-identification-in-image-main\\color_img.jpg')
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
grid_RGB =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
lower = np.array([25,150,50])
upper = np.array([35,255,255])
```

```python
mask = cv2.inRange(grid_HSV , lower , upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res = cv2.bitwise_and(grid_RGB, grid_RGB ,mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
```
